1.7 Ubicar las universidades en un mapa del Perú. Para ello, crear un mapa con las universidades públicas y distinga entre aquellas que recibieron licenciamiento o no. Luego, crear otro mapa con la misma información pero en el caso de universidades privadas.

```python
# objects from Geopandas 
import matplotlib.pyplot as plt 
import chardet
import geopandas as gpd
from geopandas import GeoSeries
from shapely.geometry import Point, LineString


# Import libraries for Folium to create heatmap
import folium
from folium import Marker, GeoJson
from folium.plugins import MarkerCluster, HeatMap

import pandas as pd
import numpy as np
import html5lib
import os
import sys
import time
import lxml

import contextily as cx  # import street map
#from pyproj import CRS
```

```python
base = pd.read_excel(r'../../Entrega_TF/Grupo2/Base_Universidad_WG2.xlsx')#leemos la base de datos exportada
base = base.drop(columns = 'Unnamed: 0' )
base
```

```python
base2 = gpd.GeoDataFrame( base, geometry=gpd.points_from_xy( base.Longitud, base.Latencia ) ) #hallamos su codifiación geográfica
base2.head(60)
```

```python
publica = base2[ base2.Tipo == 0] #Dividimos nuestra base de datos en dos
privada = base2[ base2.Tipo == 1]
```

```python
publicaL = publica[ publica.Licencia == 1] #Nuestras dos bases las subdividimos para facilitar el ploteo
publicaNL = publica[ publica.Licencia == 0]
privadaL = privada[ privada.Licencia == 1]
privadaNL = privada[ privada.Licencia == 0]
```

# Shapefile

```python
mapa = gpd.read_file(r'../../Entrega_TF/Grupo2/shape_file/DISTRITOS.shp')

mapa
```

```python
mapa.geometry
```

```python
mapa = mapa.rename({'IDDIST':'UBIGEO'}, axis =1 )
mapa = mapa.rename({'DEPARTAMEN':'DEPARTAMENTO'}, axis =1 )
mapa 
```

```python
mapa['centroid'] = mapa['geometry'].centroid
mapa
```

```python
#Universidades publicas licenciadas
ax = mapa.plot(figsize=(25, 25), color='none', edgecolor='black', zorder=1) #El mapa del Perú estará delinieado

publicaL.plot(color='blue', label = "Universidad Pública Licenciada", markersize=30, ax=ax)
publicaNL.plot(color='red', label = "Universidad Pública No Licenciada", markersize=30, ax=ax)

# to add gool map 
cx.add_basemap(ax, crs="EPSG:4326", source=cx.providers.Stamen.TerrainBackground)

plt.title('Distribución geográfica de Universidades Públicas')
plt.xticks([])
plt.yticks([])
plt.legend(loc='lower left')
```

```python
#Universidades privadas licenciadas
ax = mapa.plot(figsize=(25, 25), color='none', edgecolor='black', zorder=1)
privadaL.plot(color='blue', label = "Universidad Privada Licenciada", markersize=30, ax=ax)
privadaNL.plot(color='red', label = "Universidad Privada No Licenciada", markersize=30, ax=ax)

# to add gool map 
cx.add_basemap(ax, crs="EPSG:4326", source=cx.providers.Stamen.TerrainBackground)

plt.title('Distribución geográfica de Universidades Privadas')
plt.xticks([])
plt.yticks([])
plt.legend(loc='lower left')
```
