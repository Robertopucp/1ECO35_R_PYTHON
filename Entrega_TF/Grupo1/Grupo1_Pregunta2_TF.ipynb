##### Subir la base de datos de las manzanas en formate GeodataFrame. Crear los centroides de las manzanas.

```python
mapa1 = gpd.GeoDataFrame(mapa, geometry='geometry')
```

```python
mapa1['centroids'] = mapa1['geometry'].centroid
```

```python
mapa1['centroids']
```

```python
mapa1 = mapa1.astype({'UBIGEO': 'int'}) # pasamos la variable UBIGEO a int64 para poder hacer el merge
```

```python
manzana_data = pd.read_excel(r'../../data/Trabajo_final/Salud_mental_peru.xls')
manzana_data  # leemos el excel de salud mental para concer las ubicaciones
```

```python
manzana = gpd.GeoDataFrame(manzana_data, geometry=gpd.points_from_xy( manzana_data.long_original, manzana_data.lat_original ) )
manzana
```

```python
manzana = manzana.astype({'UBIGEO': 'int'}) # pasamos la variable UBIGEO a int64 para poder hacer el merge
```

```python
base_map_manzana = pd.merge(manzana, mapa1 , how = "inner" , on='UBIGEO')
base_map_manzana # Unimos las tablas de los distritos con el centro de salud y sus respectivas coordenadas
```

```python
data1 = gpd.GeoDataFrame({'geometry': base_map_manzana['centroids']}, crs='EPSG:4326')
data2 = gpd.GeoDataFrame({'geometry': base_map_manzana['geometry_x']}, crs='EPSG:4326')
```

```python
base_map_manzana["distancia"] = data2.apply(lambda x:data1.distance(x))
base_map_manzana # Resaltar que no corre la funcion de distance para poder conocer la distancia de los distritos con el centro de salud
```

#### Crear una variable dummy llamado Treatment que toma el valor de 1 si el centroide de la manzana se ubica a menos de 1 km de algun centro de salud mental.

```python
base_map_manzana['Treatment'] = np.where(base_map_manzana['Distancia'] < 1, 1, 0)
base_map_manzana # En caso de leerse el anterior código la variable Treatment se localiza de la siguiente manera
```

#### Crear una dummy llamado Control que toma el valor de 1 si la manzana se ubica entre 1 a 2 km del centro de salud mental.

```python
base_map_manzana['Control'] = np.where((base_map_manzana['Distancia'] <= 2) & (base_map_manzana['Distancia'] >= 1), 1, 0)
base_map_manzana # En caso de leerse el anterior código la variable Control se localiza de la siguiente manera
```
