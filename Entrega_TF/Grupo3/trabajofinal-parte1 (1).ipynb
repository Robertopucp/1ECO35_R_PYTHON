# 1. Web scrapping y Geocoding

   ## Web scrapping

### Usted debe extraer la tabla de universidades públicas y privadas. Luego debe juntar ambas.

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import re
import time
import pandas as pd
import numpy as np
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
driver1 = webdriver.Chrome( ChromeDriverManager().install() )

url = 'https://www.sunedu.gob.pe/lista-universidades/'
driver1.get( url )
driver1.maximize_window()
```

### Loops and functions

```python
all_tables = []
for i in range(2):
    unis = driver1.find_element_by_xpath('//*[@id="fw_c"]/div[1]/div/div/div/div[1]/ul/li[%d]' % (i+1))
    unis.click()
    
    #Get table of universities
    table_path = driver1.find_element_by_id('tablepress-2%d' % (i+2))
    table_html = table_path.get_attribute( 'outerHTML' )  
    
    #Read table using pandas
    table = pd.read_html(table_html)         
    print (table)
    
    table_clean = table[0].iloc[ 0:, 1: ].copy()
    #store all_tables
    all_tables.append(table_clean)
```

```python
final_data = pd.concat([all_tables[0], all_tables[1]], axis = 0)
final_data
```

### Crear una Dummy que toma el valor de 1 si la universidad es privada, y cero en universidades  públicas.

```python
final_data['Dummy_uni'] = np.where(((final_data["CONDICIÓN JURÍDICA"]== "Asociativa") | (final_data["CONDICIÓN JURÍDICA"]== "Societaria")),1,0)
final_data
```

### Crear un Dummy igual a 1 si la universidad recibió el lincencimiento de Sunedu. 

```python
final_data['dummy_lic']= np.where((final_data["DISPOSITIVO LEGAL DE LICENCIAMIENTO"].notnull()) | (final_data["DISPOSITIVO LEGAL DE LICENCIAMIENTO O DENEGATORIA"].notnull()),1,0)
final_data
```
