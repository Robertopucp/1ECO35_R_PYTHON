Crear un coefplot de las siguientes tres regresiones:

a) lbwght ~ Dummy                                                               
b) lbwght ~ Dummy + mothedu                                                               
c) lbwght~ Dummy + mothedu + lfaminc + white + Dummy:(mothedu + lfaminc + white)

Dummy : (mothedu + lfaminc + white) crea variables interactivas.

Dummy : (mothedu + lfaminc + white) = Dummy * mothedu + Dummy * lfaminc + Dummy * white

```r
#install.packages("sandwich")
library(sandwich)
```

```r
#a) lbwght ~ Dummy

mod1 <- lm(lbwght ~ dummy, data=pesos)
```

```r
mod1.est <- summary(mod1)
mod1.est
```

```r
#coeficiente estimdo de la dummy
mod1.estd <- summary(mod1)$coef["dummy",1]
mod1.estd
```

```r
#para el error estÃ¡ndar
mod1.coefs <- vcovHC(mod1, type = 'HC')
mod1.coefs
```

```r
mod1.se <- sqrt(diag(mod1.coefs))[2] # Estimated std errors, 2 porque es el segundo elemento de la diagonal
mod1.se
```

```r
library(lmtest)
library(xtable)
```

```r
#creamos la tabla
table<- matrix(0, 3, 4)
table
```

```r
#rellenamos la tabla
table[1,1]<- mod1.estd  
table[1,2]<- mod1.se   
table[2,1]<- mod2.estd
table[2,2]<- mod2.se 
table[3,1]<- mod3.estd
table[3,2]<- mod3.se 

table[1,3]<- coefci(mod1, df = Inf, vcov. = vcovHC, type = "HC")[2,1] 
table[1,4]<- coefci(mod1, df = Inf, vcov. = vcovHC, type = "HC")[2,2] 
table[2,3]<- coefci(mod2, df = Inf, vcov. = vcovHC, type = "HC")[2,1] 
table[2,4]<- coefci(mod2, df = Inf, vcov. = vcovHC, type = "HC")[2,2] 
table[3,3]<- coefci(mod3, df = Inf, vcov. = vcovHC, type = "HC")[2,1] 
table[3,4]<- coefci(mod3, df = Inf, vcov. = vcovHC, type = "HC")[2,2] 

colnames(table)<- c("Estimate","se","lower_bound","upper_bound")
rownames(table)<- c("Modelo 1", "Modelo 2", "Modelo 3")
tab<- xtable(table)
tab
```

```r
#creamos coefplot
options(repr.plot.width = 10, repr.plot.height =10)  # plot size 

ggplot(tab, aes(x=rownames(tab), y=Estimate)) +
    geom_point(size=1.5) + 
    geom_errorbar(aes(ymin=lower_bound, ymax=upper_bound) , width = 0.1,color="blue", linetype="dashed") + 
  labs(x="Linear Models", y="") + ggtitle("Confindece interval at 5% confidence level") +
  theme_bw()  + theme(text=element_text(size =12), plot.title = element_text(hjust = 0.5))
```

Use la base de datos mental_health_center para crear las columnas del tipo de instituciÃ³n del centro de salud y del RUC. NÃ³tese que esta informaciÃ³n estÃ¡ contenida en la columna InstituciÃ³n_ruc.

```r
# Abrir el documento
library("readxl")
salud <- read_excel("../../data/Centro_salud/Centro_salud_mental.xls")
salud
```

```r
#Extraer de la columna las letras de InstituciÃ³n_ruc

institucion <- c()

for (i in 1:length(salud$InstituciÃ³n_ruc)){
        w <-  gsub("[^A-Za-z]+",'', salud$InstituciÃ³n_ruc[i])
        institucion[i] <- w
        print(w)    
}
```

```r
institucion
```

```r
#Extraer de la columna los numeros de InstituciÃ³n_ruc

ruc <- c()

for (i in 1:length(salud$InstituciÃ³n_ruc)){
        x <-  gsub("[^0-9]",'', salud$InstituciÃ³n_ruc[i])
        ruc[i] <- x
        print(x)    
}
```

```r
ruc
```

```r
# AÃ±adir los dos vectos de institucion y ruc a la base de datos principal
salud <- salud %>% add_column(Institucion = institucion)
salud <- salud %>% add_column(RUC = ruc)
```
