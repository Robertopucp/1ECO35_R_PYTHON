```r
#Library

library(haven)

read_dta("C:/Users/Ariana/Desktop/ECO35/peso.dta")->datos

head(datos) 

library(ggplot2) 

```

```r
#Separar Fumadores de no fumadores

library(dplyr)

datos<-datos%>%mutate(Fuma=ifelse(cigs>0,"fumadora","No fumadora"))

```

```r
#Histograma del peso

p1<-ggplot(datos,aes(x=bwghtlbs, fill=Fuma))+
  geom_histogram(alpha=0.7,binwidth=1)+
  labs(x="Peso del Bebé (Lbs)",y="Frecuencia", title="Histograma del Peso de bebés",
       subtitle="Clasificado según madre si/no fumadora")+
  scale_fill_discrete(name="Madre")+
  theme_bw()

p2<-ggplot(datos,aes(x=bwghtlbs,fill=Fuma))+
  geom_density(alpha=.7)+theme_bw()+facet_grid(Fuma~.)+
  theme(legend.position="none")+
  labs(x="Peso del bebé (Lbs)",y="Densidad",title="Mapa de Densidad con valores de peso de bebés")

library(cowplot)
plot_grid(p2,p1,ncol=2)
```

```r
library("readxl")
salud <- read_excel("../../data/Centro_salud/Centro_salud_mental.xls")
salud
#Extraer de la columna las letras de Institución_ruc

institucion <- c()

for (i in 1:length(salud$Institución_ruc)){
  w <-  gsub("[^A-Za-z]+",'', salud$Institución_ruc[i])
  institucion[i] <- w
  print(w)    
}

print(institucion)

ruc <- c()

for (i in 1:length(salud$Institución_ruc)){
  x <-  gsub("[^0-9]",'', salud$Institución_ruc[i])
  ruc[i] <- x
  print(x)    
}

salud <- salud %>% add_column(Institucion = institucion)
salud <- salud %>% add_column(RUC = ruc)
head(salud)
```
