#### En un mismo grÃ¡fico presente la distribuciÃ³n absoluta, frecuencia y densidades de los pesos de los reciÃ©n nacidos si la madre es fumadora o no. El grÃ¡fico debe presentar titulo, tÃ­tulos en ejes y leyenda.

Crear una dummy si la madre es fumadora de cigarros

```python
def f(pesos):
    if pesos["cigs"]  > 0:
        val = 1
    else:
        val = 0
    return val

pesos['dummy'] = pesos.apply(f, axis=1)
pesos
```

```python
fig, axes = plt.subplots(1, 3)
plt.rcParams["figure.figsize"] = [15, 6]

sns.histplot(data=pesos, x="lbwght", hue="dummy", kde=True, ax=axes[2])

sns.kdeplot(data=pesos, x="lbwght", hue="dummy", cut=0, fill=True, common_norm=False, alpha=0.2, ax=axes[1])

sns.histplot(data=pesos, x="lbwght", hue="dummy", ax=axes[0])
plt.ylabel("DistribuciÃ³n absoluta")
plt.title("Log peso del bebÃ© (Si la madre es o no fumadora)")

plt.show()
```

#### Crear un coefplot de las siguientes tres regresiones:

a) lbwght ~ Dummy

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
flex = ' lbwght ~ dummy '
nocontrol_model = smf.ols( formula = flex, data = pesos )
nocontrol_est = nocontrol_model.fit().summary2().tables[1]['Coef.']['dummy']
HCV_coefs = nocontrol_model.fit().cov_HC0
nocontrol_se = np.power( HCV_coefs.diagonal() , 0.5)[1]

print(nocontrol_model.fit().summary2().tables[1])
print( f"Coefficient for OLS with controls {nocontrol_est}" )
# print unconditional effect of gender and the corresponding standard error
print( f'The estimated gender coefficient is {nocontrol_est} and the corresponding robust standard error is {nocontrol_se}' )

nocontrol_model.fit().conf_int( alpha=0.05 ).loc['dummy']
```

b) lbwght ~ Dummy + mothedu

```python
flex1 = ' lbwght ~ dummy + motheduc '
control_model1 = smf.ols( formula = flex1, data = pesos) 
control_est1 = control_model1.fit().summary2().tables[1]['Coef.']['dummy']

print(control_model1.fit().summary2().tables[1])
print( f"Coefficient for OLS with controls {control_est1}" )

HCV_coefs1 = control_model1.fit().cov_HC0
control_se1 = np.power( HCV_coefs1.diagonal() , 0.5)[1]

control_model1.fit().conf_int( alpha=0.05 ).loc['dummy']
```

c) lbwght~ Dummy + mothedu + lfaminc + white + Dummy:(mothedu + lfaminc + white)

```python
flex2 = ' lbwght ~ dummy + motheduc + lfaminc + white + (dummy)*( motheduc +  lfaminc + white) '
control_model2 = smf.ols( formula = flex2, data = pesos) 
control_est2 = control_model2.fit().summary2().tables[1]['Coef.']['dummy']

print(control_model2.fit().summary2().tables[1])
print( f"Coefficient for OLS with controls {control_est2}" )

HCV_coefs2 = control_model2.fit().cov_HC0
control_se2 = np.power( HCV_coefs2.diagonal() , 0.5)[1]

control_model2.fit().conf_int( alpha=0.05 ).loc['dummy']
```

```python
table3 = np.zeros( ( 3 , 4 ) )

table3[0,0] = nocontrol_est
table3[0,1] = nocontrol_se 
table3[0,2] = nocontrol_model.fit().conf_int( alpha=0.05 ).loc['dummy'][0]
table3[0,3] = nocontrol_model.fit().conf_int( alpha=0.05 ).loc['dummy'][1] 

table3[1,0] = control_est1
table3[1,1] = control_se1  
table3[1,2] = control_model1.fit().conf_int( alpha=0.05 ).loc['dummy'][0]
table3[1,3] = control_model1.fit().conf_int( alpha=0.05 ).loc['dummy'][1] 

table3[2,0] = control_est2
table3[2,1] = control_se2  
table3[2,2] = control_model2.fit().conf_int( alpha=0.05 ).loc['dummy'][0]
table3[2,3] = control_model2.fit().conf_int( alpha=0.05 ).loc['dummy'][1] 

table3_pandas = pd.DataFrame( table3, columns = [ "Estimate","Std. Error","Lower_bound" , "Upper_bound"])
table3_pandas.index = [ "no_control", "lbwght ~ dummy + mothedu", "lbwght ~ dummy + mothedu + ..."]

table3_pandas.reset_index(inplace = True)
table3_pandas.rename(columns = {"index" : "Model"}, inplace = True)

table3_pandas.round(8)
```

```python
table3_pandas['errors'] = table3_pandas['Estimate'] - table3_pandas['Lower_bound']
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
    
    
   # Coefficients
ax.scatter(x=pd.np.arange(table3_pandas.shape[0]), 
         marker='o', s=80, 
         y=table3_pandas['Estimate'], color='black')

# Error bars for 95% confidence interval
# Can increase capsize to add whiskers
eb1 = plt.errorbar(x=table3_pandas["Model"], y=table3_pandas['Estimate'],
            yerr=table3_pandas['errors'], color = 'darkblue', ls='', capsize = 8)

eb1[-1][0].set_linestyle('--')
    
# Set title & labels
plt.title('Coefficients confindece interval at 5% confidence level',fontsize=15)
ax.set_ylabel('Coefficients',fontsize=15)
ax.set_xlabel('',fontsize=22)

    
plt.show()
```
