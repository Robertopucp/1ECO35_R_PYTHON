#### Filtramos las variables de d1 y eliminamos duplicados

#### Quitamos el espacio en cada variable separada por años 

```r
names <- names(d1)
lista <- c()

for (i in 1:length(names)){
        h <-  gsub('[/_]','', names[i])
        lista[i] <- h
        print(h)    
}
```

```r
main_var <- c("numper","numpanh18", "mes18", "ubigeo18", "dominio18", "p400a318", "p402218")

d1 <- d1[,(names(d1) %in% main_var)]
d1
```

```r
d1 <- d1 %>% rename ("numpanh" = "numpanh18",
       "mes" = "mes18",
       "ubigeo" = "ubigeo18",
       "dominio" = "dominio18",
       "p400a3" = "p400a318",
       "p4022" = "p402218",
       )  
d1
```

## MERGE

```r
finalbd <- merge( d1, d2, by = c("ubigeo"), all.x = TRUE)
head(finalbd)
```

### Labels 

```r
var_lab(finalbd$numper) = "Unique individual identifier"
```

```r
var_lab(finalbd$numpanh) = "Unique household identifier in each year"
```

```r
var_lab(finalbd$mes) = "Month of the survey"
```

```r
var_lab(finalbd$p400a3) = "Year of birth of the person, reported in each year"
```

```r
var_lab(finalbd$p4022) = "Dummy : 1 if the person was sick in the last 4 weeks"
```

```r
var_lab(finalbd$unidos) = "Dummy : 1 if Programa Juntos was applied in the district"
```

### Porcentaje de distritos por departamentos que se beneficiaron del Programa Juntos en cada año

```r
finalbd2 <- finalbd %>% group_by(year, region) %>% summarize(cantidad = n(), cantidad_juntos=sum(unidos))
finalbd2$percent_juntos_depart <- finalbd2$cantidad_juntos *100 / finalbd2$cantidad
finalbd2
```
