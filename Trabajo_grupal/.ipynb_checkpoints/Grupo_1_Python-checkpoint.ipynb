```python
import pandas as pd
import numpy as np
import chardet # to get string character format 
import re
```

```python
#Data enaho panel (2015-2019)
d1 = pd.read_stata("../data/Juntos_program/data.dta")
#Data Programa Juntos
d2 = pd.read_stata("../data/Juntos_program/unidos.dta")
```

##### Replicar DHS y limpiar base de datos

```python
d1.columns      # to see de variables
```

```python
filter_list = []  

for i in range(0,len(list(d1.columns))):

        w =  re.sub('[\_]','', list(d1.columns)[i])   # filter the variables with "_"
        filter_list.append(w)
```

```python
filter_list1 = []
for i in range(0,len(list(filter_list))):

        w =  re.sub(".{2}$",'', list(filter_list)[i]) # filter without the last 2 characters
        filter_list1.append(w)
```

```python
# drop duplicates in a list
nuevo = list(dict.fromkeys(filter_list1))
nuevo
```

##### Base de datos tipo panel

```python
dat_panel = pd.wide_to_long(d1, stubnames = nuevo, i = ['numper'] , j = 'year' , sep = '_').reset_index()
```

```python
# add the variable 'year' and get the data panel
nuevo[0] = d1.columns[0]
nuevo[1] = d1.columns[1]
nuevo.append("year")
dato_panel = dat_panel[nuevo]
dato_panel = dato_panel.rename(columns={'numpanh15': 'numpanh'})
dato_panel        
```

```python
# merge the 2 datas
total = pd.merge(dato_panel,d2,how='inner', on='ubigeo') # we use ubigeo to select all the people in both data
total
```
