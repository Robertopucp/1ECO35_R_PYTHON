```python
import pandas as pd
import numpy as np
import re
```

```python
#Data ENAHO PANEL 2015-2019
d1 = pd.read_stata("../data/Juntos_program/data.dta")
#Data Programa Juntos
d2 = pd.read_stata("../data/Juntos_program/unidos.dta")
```

```python
filter_list = []
filter_list1 = []

for i in range(0,len(list(d1.columns))):

        z =  re.sub('[\_]','', list(d1.columns)[i])
        filter_list.append(z)

for i in range(0,len(list(filter_list))):

        z =  re.sub(".{2}$",'', list(filter_list)[i])
        filter_list1.append(z)
```

```python
nuevo = list(dict.fromkeys(filter_list1))
nuevo
```

##### Datos tipo panel

```python
dat_panel = pd.wide_to_long(d1, stubnames = nuevo, i = ['numper'] , j = 'year' , sep = '_').reset_index()
```

```python
nuevo[0] = d1.columns[0]
nuevo[1] = d1.columns[1]
nuevo.append("year")
dato_panel = dat_panel[nuevo]
dato_panel = dato_panel.rename(columns={'numpanh15': 'numpanh'})
dato_panel        
```

```python
total = pd.merge(dato_panel,d2,on='ubigeo')
total
```
