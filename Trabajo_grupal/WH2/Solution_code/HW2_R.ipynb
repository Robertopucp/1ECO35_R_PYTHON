```r
notas= read.csv("../../../data/notas.csv")  # relative path 
```

```r


i <- 1


nrows <- dim(notas)[1]


for (i in 1:nrows){
    
    if (notas$notas[i] >= 11){
        notas$Estado[i] = "Ingresante"
        
    } else {
        notas$Estado[i] = "No ingresó"
    }
       
    
    if (notas$carrera[i] == "a001"){
        
        notas$Nombre_carrera[i] = "Economía"
        
    } else if (notas$carrera[i] == "a002"){

            notas$Nombre_carrera[i] = "Química"
    } else {
        
        notas$Nombre_carrera[i] = "Filosofia"
        
    }
 
    i <- i + 1 

    
}
      
```
