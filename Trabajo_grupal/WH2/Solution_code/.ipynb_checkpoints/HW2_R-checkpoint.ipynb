### 1.1 If statement

```r
nrows <- dim(notas)[1]


for (i in 1:nrows){
    
    if (notas$notas[i] >= 11){
        notas$Estado[i] <- "Ingresante"
        
    } else {
        notas$Estado[i] <- "No ingresó"
    }
       
    
    if (notas$carrera[i] == "a001"){
        
        notas$Nombre_carrera[i] <- "Economía"
        
    } else if (notas$carrera[i] == "a002"){

        notas$Nombre_carrera[i] <- "Química"
        
    } else if (notas$carrera[i] == "a003") {
        
        notas$Nombre_carrera[i] <- "Filosofia"
        
    } else{
        notas$Nombre_carrera[i] <- "NA"
    }

    
}
      
```

### 2.1 Coefficients loop

```r
set.seed(756)

x1 <- runif(10000)
x2 <- runif(10000)
x3 <- runif(10000)
x4 <- runif(10000)

e <- rnorm(10000)

# Poblacional regression (Data Generating Process GDP)

Y <- 1 + 0.8*x1 + 1.2*x2 + 0.5*x3 + 1.5*x4 + e
```

```r
sample_size <- c(10,50,100,200,500,1000,5000)
betas <- c()
for (j in sample_size){
    
    X <- cbind(matrix(1,j), x1[1:j],x2[1:j],x3[1:j],x4[1:j])
    Y1 = Y[1:j]
    beta <- solve(t(X) %*% X) %*% (t(X) %*% Y1)
    betas <- append(betas,c(beta))
}
```

## 3.0 Function section

```r
set.seed(1246)

M <- matrix(runif(1000), ncol = 10)
```

```r
dim(M)
```

```r
standarize <- function(X){
    Z <- matrix(0,100,10)
    
    for (j in seq( 1, dim(Z)[2] ) ){
        n <- dim(Z)[1]
        a <- mean(X[,j])
        b <- sd(X[,j])*sqrt((n-1)/n)   # standar deviation correction 
        Z[,j] <- (X[,j]-a)/b
        
    }
    
    return(Z)
}
```

```r
head(standarize(M))
```
