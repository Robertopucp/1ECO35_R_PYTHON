##### Pregunta 8: Crear columnas con las siguiente información: el % de mujeres que no escriben ni leen, % de varones que no escriben ni leen y % de nativos respecto al total de la población.

```r
junin$Total_poblacion = junin$peruvian_men + junin$peruvian_women + junin$foreign_men + junin$foreign_women
```

```r
junin$Porc_women_nread_nwrite = (junin$no_instruction_women * 100) / junin$Total_poblacion
```

```r
junin$Porc_men_nread_nwrite = (junin$no_instruction_men * 100) / junin$Total_poblacion
```

```r
junin$Porc_natives = (junin$natives * 100) / junin$Total_poblacion
```
